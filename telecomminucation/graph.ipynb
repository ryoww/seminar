# $S_p = Ge^{-2G}$の描画

```python
import numpy as np
import matplotlib.pyplot as plt

def funcp(G):
    return G * np.exp(-2 * G)

# Gの値の範囲を指定
G = np.linspace(0, 10, 100)

# 関数を計算
S_p_p = funcp(G)

# グラフを描画
plt.plot(G, S_p_p, label= "Pure ALOHA")
plt.xlabel("Traffic(G)")
plt.ylabel("Throughput(S_p)")
plt.title("Function: S_p = Ge^(-2G)")
plt.grid(True)
plt.legend()
plt.show()
```

# $S_p = Ge^{-G}$の描画

```python
import numpy as np
import matplotlib.pyplot as plt

def funcs(G):
    return G * np.exp(-1 * G)

# Gの値の範囲を指定
G = np.linspace(0, 10, 100)

# 関数を計算
S_p_s = funcs(G)

# グラフを描画
plt.plot(G, S_p_s, color= "darkorange", label= "Slotted ALOHA")
plt.xlabel('Traffic(G)')
plt.ylabel('Throughput(S_p)')
plt.title('Function: S_p = Ge^(-G)')
plt.grid(True)
plt.legend()
plt.show()
```

# Pure ALOHA, Slotted ALOHAの比較

```python
import numpy as np
import matplotlib.pyplot as plt

def funcp(G):
    return G * np.exp(-2 * G)


def funcs(G):
    return G * np.exp(-1 * G)

# Gの値の範囲を指定
G = np.linspace(0, 10, 100)

# 関数を計算
S_p_p = funcp(G)
S_p_s = funcs(G)

# グラフを描画
plt.plot(G, S_p_p, label= "Pure ALOHA")
plt.plot(G, S_p_s, label= "Slotted ALOHA")
plt.xlabel('Traffic(G)')
plt.ylabel('Throughput(S_p)')
plt.title('Function: S_p = Ge^(-2G) & S_p = Ge^(-G)')
plt.grid(True)
plt.legend()
plt.show()
```
