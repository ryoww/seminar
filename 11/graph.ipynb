# $L_{fs} = 20log_{10} f + 20log_{10} d + 20log_{10} \frac{4 \pi}{c}$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ["svg"]

def calculate_L_fs(d, f):
    c = 3.0e8  #  [m/s]
    Lfs = 20 * np.log10(f) + 20 * np.log10(d) + 20 * np.log10((4 * np.pi) / c)
    return Lfs

# 
d = np.linspace(0, 100, 10000)  #  d [m] 
f_5 = 5.2e9  #  f [Hz]
f_25 = 2.45e9 #  f [Hz]

# L_fs 
L_fs_52g = calculate_L_fs(d, f_5)
L_fs_25g = calculate_L_fs(d, f_25)

# 
plt.plot(d, L_fs_52g, label= "5.2 [GHz]")
plt.plot(d, L_fs_25g, label= "2.45[GHz]")
plt.ylim(40,90)
plt.xlim(0,100)
plt.gca().invert_yaxis()
plt.xlabel(r'd [m]')
plt.ylabel(r'$L_{fs}$[dB]')
plt.title(r'Plot of $L_{fs} = 20\log_{10}(f) + 20\log_{10}(d) + 20\log_{10}(4 \pi/c))$')
plt.grid(True)
plt.legend()
plt.show()
```

# $L = 20 \cdot \log_{10} f + N \cdot \log_{10} d + L_f (n) - 28$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ["svg"]

def calculate_L(d, N, f, Lf, n):
    L = 20 * np.log10(f) + N * np.log10(d) + Lf * n - 28
    return L

d = np.linspace(0, 50, 1000)  #  d [m]

#  5.2[GHz] 
L_syu_5 = calculate_L(d, 30, 5.2e3, 0, 0)
# / 2.45[GHz] 
L_syu_25 = calculate_L(d, 28, 2.45e3, 0, 0)
#  5.2[GHz] 
L_ko_5 = calculate_L(d, 28, 5.2e3, 0, 0)
#  5.2[GHz] 
L_offi_5 = calculate_L(d, 31, 5.2e3, 0, 0)
#  2.45[GHz] 
L_offi_25 = calculate_L(d, 30, 2.45e3, 0, 0)




# 
plt.plot(d, L_syu_5, label= "5.2[GHz]")
plt.plot(d, L_syu_25, label= "/2.45[GHz]")
plt.plot(d, L_ko_5, label= "5.2[GHz]")
plt.plot(d, L_offi_5, label= "5.2[GHz]")
plt.plot(d, L_offi_25, label= "2.45[GHz]")
plt.ylim(60,100)
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.xlabel('d [m]')
plt.ylabel('L [dB]')
plt.title(r'Plot of $L = 20\log_{10}(f) + N\log_{10}(d) + L_f(n) - 28$')
plt.text(x=22, y=74.5, s=r"$L_f$ = 0 [dB]")
plt.grid(True)
plt.legend(prop={"family":"MS Gothic"})
plt.show()
```
