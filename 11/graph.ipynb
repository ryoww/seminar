# $L_{fs} = 20log_{10} f + 20log_{10} d + 20log_{10} \frac{4 \pi}{c} + 120$

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_L_fs(d, f):
    c = 3.0e8  # 光速 [m/s]
    Lfs = 20 * np.log10(f) + 20 * np.log10(d) + 20 * np.log10(4 * np.pi / c) + 120
    return Lfs

# パラメータの設定
d = np.linspace(0, 100, 1000)  # 距離 d [m] の範囲と分割数
f = 2.45e9  # 周波数 f [Hz]

# L_fs の計算
L_fs = calculate_L_fs(d, f)

# グラフの描画
plt.plot(d, L_fs)
plt.gca().invert_yaxis()
plt.xlabel('d [m]')
plt.ylabel('L_fs')
plt.title('Plot of L_fs = 20*log10(f) + 20*log10(d) + 20*log10(4*pi/c) + 120')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_L_fs(d, f):
    c = 3.0e8  # 光速 [m/s]
    Lfs = 20 * np.log10(f) + 20 * np.log10(d) + 20 * np.log10((4 * np.pi) / c)
    return Lfs

# パラメータの設定
d = np.linspace(0, 100, 10000)  # 距離 d [m] の範囲と分割数
f_5 = 5.2e9  # 周波数 f [Hz]
f_25 = 2.45e9 # 周波数 f [Hz]

# L_fs の計算
L_fs_5g = calculate_L_fs(d, f_5)
L_fs_25g = calculate_L_fs(d, f_25)

# グラフの描画
plt.plot(d, L_fs_5g, label= "5[GHz]")
plt.plot(d, L_fs_25g, label= "2.45[GHz]")
plt.ylim(40,90)
plt.xlim(0,100)
plt.gca().invert_yaxis()
plt.xlabel(r'd [m]')
plt.ylabel(r'$L_{fs}$[dB]')
plt.title(r'Plot of $L_{fs} = 20log_{10}(f) + 20log_{10}(d) + 20log_{10}(4 \pi/c))$')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_L(d, N, f, Lf, n):  # 床・天井・壁を通過することによる負荷損失
    L = 20 * np.log10(f) + N * np.log10(d) + Lf * n - 28
    return L

# パラメータの設定
L_f = 0
d = np.linspace(1, 50, 1000)  # 距離 d [m]
N = 0.2  # 減衰係数 N
f = 5.2 * (10 ** 9)  # 周波数 f の範囲と分割数
n = 2  # 通過する床等の枚数 n

# L の計算
L = calculate_L(d, N, f, L_f, n)

# グラフの描画
plt.plot(f, L)
plt.xlabel('f')
plt.ylabel('L')
plt.title('Plot of L = 20*log10(f) + N*log10(d) + Lf(n) - 28')
plt.grid(True)
plt.show()
```
