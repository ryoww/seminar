# $L_{fs} = 20\log_{10} f + 20\log_{10} d + 20\log_{10} \frac{4 \pi}{c}$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ["svg"]

def calculate_L_fs(d, f):
    c = 3.0e8  # 光速 [m/s]
    Lfs = 20 * np.log10(f) + 20 * np.log10(d) + 20 * np.log10((4 * np.pi) / c)
    return Lfs

# パラメータの設定
d = np.linspace(0, 100, 10000)  # 距離 d [m] の範囲と分割数
f_5 = 5.2e9  # 周波数 f [Hz]
f_25 = 2.45e9 # 周波数 f [Hz]

# L_fs の計算
L_fs_52g = calculate_L_fs(d, f_5)
L_fs_25g = calculate_L_fs(d, f_25)

# グラフの描画
plt.plot(d, L_fs_52g, label= "5.2 [GHz]")
plt.plot(d, L_fs_25g, label= "2.45[GHz]")
plt.ylim(40,90)
plt.xlim(0,100)
plt.gca().invert_yaxis()
plt.xlabel(r'd [m]')
plt.ylabel(r'$L_{fs}$[dB]')
plt.title(r'Plot of $L_{fs} = 20\log_{10}(f) + 20\log_{10}(d) + 20\log_{10}(4 \pi/c))$')
plt.grid(True)
plt.legend()
plt.show()
```

# $L = 20 \cdot \log_{10} f + N \cdot \log_{10} d + L_f (n) - 28$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ["svg"]

def calculate_L(d, N, f, Lf, n):
    L = 20 * np.log10(f) + N * np.log10(d) + Lf * n - 28
    return L

d = np.linspace(0, 50, 1000)  # 距離 d [m]

# 集合住宅内 5.2[GHz] の計算
L_syu_5 = calculate_L(d, 30, 5.2e3, 0, 0)
# 集合住宅内/戸建て住宅内 2.45[GHz] の計算
L_syu_25 = calculate_L(d, 28, 2.45e3, 0, 0)
# 戸建て住宅内 5.2[GHz] の計算
L_ko_5 = calculate_L(d, 28, 5.2e3, 0, 0)
# オフィス内 5.2[GHz] の計算
L_offi_5 = calculate_L(d, 31, 5.2e3, 0, 0)
# オフィス内 2.45[GHz] の計算
L_offi_25 = calculate_L(d, 30, 2.45e3, 0, 0)




# グラフの描画
plt.plot(d, L_syu_5, label= "集合住宅内5.2[GHz]")
plt.plot(d, L_syu_25, label= "集合住宅内/戸建て住宅内2.45[GHz]")
plt.plot(d, L_ko_5, label= "戸建て住宅内5.2[GHz]")
plt.plot(d, L_offi_5, label= "オフィス内5.2[GHz]")
plt.plot(d, L_offi_25, label= "オフィス内2.45[GHz]")
plt.ylim(60,100)
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.xlabel('d [m]')
plt.ylabel('L [dB]')
plt.title(r'Plot of $L = 20\log_{10}(f) + N\log_{10}(d) + L_f(n) - 28$')
plt.text(x=22, y=74.5, s=r"$L_f$ = 0 [dB]")
plt.grid(True)
plt.legend(prop={"family":"MS Gothic"})
plt.show()
```

# 指数分布 $p(x) = \lambda e^{- \lambda x} \quad (\because \lambda = \frac{1}{\sigma})$

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ["svg"]

# 標準偏差を設定
std_dev = 30*10**(-9)

# レートパラメータを計算
lambda_param = (1 / std_dev)

# x軸の値を生成
# x = np.linspace(0, 300e-9, 100000)
x = np.linspace(std_dev, std_dev*10, 100000)

# 指数分布の確率密度関数を計算
pdf = lambda_param * np.exp(-lambda_param * x)
pdf_wong = np.exp(-lambda_param * x)

# グラフを描画
plt.figure()
# plt.plot(x, pdf, label= r"$p(x) = \lambda e^{- \lambda x}$")
plt.plot(x,pdf_wong, label= r"$p(x) = e^{- \lambda x}$")
plt.title('指数分布の計算例',fontname = 'MS Gothic')
plt.xlabel('相対遅延時間',fontname = 'MS Gothic')
plt.ylabel('分布率例',fontname = 'MS Gothic')
plt.legend()
plt.xlim(0,300e-9)
# plt.ylim(0,0.04)
plt.grid()
plt.legend()
plt.show()
```

# ラプラシアン分布

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ["svg"]

# 標準偏差を設定（単位をラジアンに変換）
std_dev_degrees = 30
# std_dev_radians = np.radians(std_dev_degrees)

# スケールパラメータを計算
scale_param = std_dev_degrees / np.sqrt(2)

# x軸の値を生成
x = np.linspace(-180, 180, 500)

# ラプラシアン分布の確率密度関数を計算
pdf = (1 / (2 * scale_param)) * np.exp(-np.abs(x) / scale_param)

# グラフを描画
plt.figure(figsize=(8, 6))
plt.plot(x, pdf)
plt.title('Laplacian Distribution Approximation')
plt.xlabel('x (Degrees)')
plt.ylabel('Probability Density')
plt.xlim(-180,180)
plt.legend()
plt.grid()
plt.show()
```
