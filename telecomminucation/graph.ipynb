# $S_p = Ge^{-2G}$の描画

```python
import numpy as np
import matplotlib.pyplot as plt

def funcp(G):
    return G * np.exp(-2 * G)

# Gの値の範囲を指定
G = np.linspace(0, 10, 1000)

# 関数を計算
S_p_p = funcp(G)

# 関数の最大値を出力
print(S_p_p.max()*100, "%")

# グラフを描画
plt.plot(G, S_p_p, label= "Pure ALOHA")
plt.xlabel("Traffic(G)")
plt.ylabel("Throughput(S_p)")
plt.title("Function: S_p = Ge^(-2G)")
plt.grid(True)
plt.legend()
plt.show()
```

# $S_p = Ge^{-G}$の描画

```python
import numpy as np
import matplotlib.pyplot as plt

def funcs(G):
    return G * np.exp(-1 * G)

# Gの値の範囲を指定
G = np.linspace(0, 10, 1000)

# 関数を計算
S_p_s = funcs(G)

# 関数の最大値を出力
print(S_p_s.max()*100, "%")

# グラフを描画
plt.plot(G, S_p_s, color= "darkorange", label= "Slotted ALOHA")
plt.xlabel('Traffic(G)')
plt.ylabel('Throughput(S_p)')
plt.title('Function: S_p = Ge^(-G)')
plt.grid(True)
plt.legend()
plt.show()
```

# Pure ALOHA, Slotted ALOHAの比較

```python
import numpy as np
import matplotlib.pyplot as plt

def funcp(G):
    return G * np.exp(-2 * G)


def funcs(G):
    return G * np.exp(-1 * G)

# Gの値の範囲を指定
G = np.linspace(0, 10, 1000)

# 関数を計算
S_p_p = funcp(G)
S_p_s = funcs(G)

# グラフを描画
plt.plot(G, S_p_p, label= "Pure ALOHA")
plt.plot(G, S_p_s, label= "Slotted ALOHA")
plt.xlabel('Traffic(G)')
plt.ylabel('Throughput(S_p)')
plt.title('Function: S_p = Ge^(-2G) & S_p = Ge^(-G)')
plt.grid(True)
plt.legend()
plt.show()
```

# $S_N = \frac{Ge^{-aG}}{G(1 + 2a) + e^{-aG}}$の描画

```python
import numpy as np
import matplotlib.pyplot as plt

def func(G, a):
    numerator = G * np.exp(-a * G)
    denominator = G * (1 + 2 * a) + np.exp(-a * G)
    return numerator / denominator

# パラメータの設定
G = np.linspace(0.01, 100, 15000)  # Gの範囲と分割数を指定

# 関数の評価
S_N_1 = func(G, 1)
S_N_5 = func(G, 0.5)
S_N_0 = func(G, 0)

# グラフの描画
plt.plot(G, S_N_1, label= "a = 1")
plt.plot(G, S_N_5, label= "a = 0.5")
plt.plot(G, S_N_0, label= "a = 0")

plt.xscale("log")
plt.xlabel('G')
plt.ylabel('S_N')
plt.title('Function S_N')
plt.grid(True)
plt.legend()
plt.show()
```

# $S_P = \frac{G(1 + G + aG(1 + G + aG /2))e^{-G(1+2a)}}{G(1 + 2a) - (1 - e^{-aG}) + (1 + aG)e^{-G(1 + a)}}$の描画

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_S_P(G, a):
    numerator = G * (1 + G + a * G * (1 + G + (a * G) / 2)) * np.exp(-G * (1 + 2 * a))
    denominator = G * (1 + 2 * a) - (1 - np.exp(-a * G)) + (1 + a * G) * np.exp(-G * (1 + a))
    S_P = numerator / denominator
    return S_P

# パラメータの設定
G = np.linspace(0.01, 100, 15000)  # Gの範囲と分割数を指定

# 関数の評価
S_N_1 = calculate_S_P(G, 1)
S_N_5 = calculate_S_P(G, 0.5)
S_N_0 = calculate_S_P(G, 0)

# グラフの描画
plt.plot(G, S_N_1, label= "a = 1")
plt.plot(G, S_N_5, label= "a = 0.5")
plt.plot(G, S_N_0, label= "a = 0")

plt.xscale("log")
plt.xlabel('G')
plt.ylabel('S_P')
plt.title('Function S_P')
plt.grid(True)
plt.legend()
plt.show()
```
