```python
import numpy as np
import matplotlib.pyplot as plt

def funcp(G):
    return G * np.exp(-2 * G)


def funcs(G):
    return G * np.exp(-1 * G)

# G
G = np.linspace(0, 10, 100)

# 
S_p_p = funcp(G)
S_p_s = funcs(G)

# 
plt.plot(G, S_p_p, label= "Pure ALOHA")
plt.plot(G, S_p_s, label= "Slotted ALOHA")
plt.xlabel('Traffic(G)')
plt.ylabel('Throughput(S_p)')
plt.title('Function: S_p = Ge^(-2G) & S_p = Ge^(-G)')
plt.grid(True)
plt.legend()
plt.show()
```
